```python
# Importing NumPy and setting up the dataset

import numpy as np

data = ([
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
])
```

```python
# Step 1 (Encoding of Strings)
# Encoding the strings using Label Encoding

for sample in data:
  if sample[3] == 'Apple':
    sample[3] = 0
  elif sample[3] == 'Banana':
    sample[3] = 1
  elif sample[3] == 'Orange':
    sample[3] = 2

print("Data with Encoded Strings:")
print(data)
print()

# Separating the data into the feature matrix 'X' and the label vector 'y'

data = np.array(data)
X = data[:, [0, 1, 2]].copy()
y = data[:, 3].copy()
print("Feature Matrix X:")
print(X)
print()
print("Label Vector y:")
print(y)
```

```python
# Step 2 (Euclidean Distance Function)

def euc_dist(p1, p2, n):
  square_dist = 0
  for i in range(n):
    square_dist += (p1[i]- p2[i])**2
  return np.sqrt(square_dist)
```

```python
# Step 3 (Implementing the KNN Classifier)

class KNN:
  # Initializing
  def __init__(self, k=3):
    self.k = k

  # Storing the training data
  def fit(self, X, y):
    self.X_train = X
    self.y_train = y

  # Predicting the label for a single data point
  def predict_one(self, x):
    distances_with_label = []
    # Calculating the (Euclidean) distance from x to all the points in the training set
    for i in range(len(self.X_train)):
      distances_with_label.append([euc_dist(x, self.X_train[i], 3), self.y_train[i]])

    # Finding the k nearest neighbours and voting on the prediction
    distances_with_label.sort(key=lambda x:x[0])

    votes = [0]*3
    for i in range(self.k):
      votes[int(distances_with_label[i][1])] += 1
    return votes.index(max(votes))

  # Predicting the labels for a test set
  def predict(self, X_test):
    predictions = []
    for x in X_test:
      predictions.append(self.predict_one(x))
    return predictions
```

```python
# Step 4 (Testing the Classifier)

test_data = np.array([
    [118, 6.2, 0],
    [160, 7.3, 1],
    [185, 7.7, 2]
])

classifier = KNN(k=3)
classifier.fit(X, y)
y_pred = classifier.predict(test_data)

print("Predicted labels for the test data:")
print(y_pred)
```

```python
# Step 5 (Evaluation)

# Evaluating for the original predictions with k = 3
if y_pred == [1, 0, 2]:
  print("Classifier predicted the labels accurately")
else:
  print("Classifier did not predict the labels accurately")
print()

# Evaluating for different values of k like 1, 3 and 5
check_k = [1, 3, 5]
for k in check_k:
  classifier = KNN(k=k)
  classifier.fit(X, y)
  y_pred = classifier.predict(test_data)

  print(f"For k = {k}:")
  print(y_pred)
  if y_pred == [1, 0, 2]:
    print("Classifier predicted the labels accurately")
  else:
    print("Classifier did not predict the labels accurately")
  print()
```
